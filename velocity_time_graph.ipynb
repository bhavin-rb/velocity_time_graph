```python
import matplotlib.pyplot as plt
```

```python
time_velocity = zip([0,5,10,15,20,25,30,35,40], [0,10,20,20,20,15,10,5,0])
#print(*time_distance)
x, y = zip(*time_velocity) # This is like calling zip((0, 0), (5, 10), (10, 15, (15, 20), (20, 20), (25, 20), (30, 15), (35, 5), (40, 0))
print(x)
print(y)
```

```python
plt.figure()
plt.plot(x,y)
#plt.xticks(x)
```

```python
plt.gca().fill_between(x,y)
```

```python
plt.xlabel('Time(seconds)')
# add a label to the y axis
plt.ylabel('Velocity(m/s)')
# add a title
plt.title('Velocity-time Graph of Uniform Motion')
```

Acceleration is change in velocity over time. In other words it is the slope of the the triangle formed between time 0s and 10s, having vertices (height) of 20.


<img src="acceleration equation.png">

From the plotted graph, v0 = 0 m/s, v1 = 20 m/s, t0 = 0s and t1 = 10s. Hence acceleration is given by:


<img src="acceleration equation1.png">

```python
# lets define our variables
v0 = 0
t0 = 0
v1 = 20
t1 = 10

# for printing superscript for S.I Unit of acceleration
SUP = str.maketrans("0123456789", "⁰¹²³⁴⁵⁶⁷⁸⁹")

# acceleration, a formula (equation)
a = (v1 - v0)/(t1 - t0)

f"Acceleration of the car is {a}" +" "f"m/s2".translate(SUP)
```

Deceleration is negative acceleration, when the car is coming to a stop.  From the plot (graph), the car is slowing down from 20s and comes to a stop on 40s.  That means we have negative slop formed between triangle 20s and 40s havinf vertices (height) of 20.

We use the same formula of acceleration.

From the plotted graph, v0 = 20 m/s, v1 = 0 m/s, t0 = 20s and t1 = 40s.
The initial velocity, v0 is 20m/s and the final velocity, v1 is 0m/s as the car comes to a stop. This is between period of time,20s and 40s.

```python
# lets define our variables
v0 = 20
t0 = 20
v1 = 0
t1 = 40

# acceleration, a formula (equation)
a = (v1 - v0)/(t1 - t0)

f"Acceleration of t|he car is {a}" +" "f"m/s2".translate(SUP)
```

```python
print("Deceleration = negative acceleration")
```

Total Ditsnace travelled by a Car is equal to the area under the graph. The plot (graph) is a Trapezium and area of Trapezium is given by:


<img src="Area of Trapezium.png">

```python
# lets define our variables
a = 20 - 10  # from 10s to 20s
b = 40 - 0   # from 0s to 40s
h = 20

area = ((a + b) * h)/2

f"Total distance travelled by the car is {area} m"
```
